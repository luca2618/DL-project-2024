```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.DataFrame([
    ['Bio Benchmarks', 0.57684, 0.59328, 0.47244, 0.61830, 0.56969, 0.49321, 0.54682],
    ['Math Benchmarks', 0.35791, 0.35951, 0.26877, 0.36252, 0.34332, 0.30692, 0.34000],
    ['All Benchmarks', 0.47131, 0.48073, 0.37566, 0.49500, 0.46057, 0.40395, 0.44720],
    ], 
    columns=['Accuracy Metric', 'Mistral 7B Instruct', 'BioMistral 7B', 'MetaMath 7B', 'MetaBio-Merged 7B', 'LoRA-Bio 7B', 'LoRA-Math 7B', 'LoRA-Merged 7B']
)
dataframe
```

```python
# Reshape the DataFrame to long format
long_df = dataframe.melt(id_vars='Accuracy Metric', 
                         var_name='Model', 
                         value_name='Accuracy')

# Plotting the grouped bar chart
plt.figure(figsize=(12, 6))
sns.barplot(
    data=long_df, 
    x='Accuracy Metric', 
    y='Accuracy', 
    hue='Model',
    palette="muted" # "colorblind" # ["#b30000", "#7c1158", "#4421af", "#1a53ff", "#0d88e6", "#00b7c7", "#5ad45a"] # ["#fd7f6f", "#7eb0d5", "#b2e061", "#bd7ebe", "#ffb55a", "#ffee65", "#beb9db"]
)

# Customize the plot
plt.title("Accuracy Across Benchmarks for Various Models", fontsize=19)
plt.xlabel("Accuracy Metric", fontsize=15)
plt.ylabel("Accuracy", fontsize=15)
plt.xticks(rotation=0, ha='right', fontsize=14)
plt.legend(title='Model', title_fontsize=20, bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=18)
plt.tight_layout()

# Show the plot
plt.show()
```
